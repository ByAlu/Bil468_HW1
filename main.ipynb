```python
import TemplateMatcher

matcher = TemplateMatcher.TemplateMatcher()

thresholds = [0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9]
method_thresholds = [0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9]

results = matcher.train(thresholds, method_thresholds)

print(results)

result = matcher.evaluate()

print(result)
```
